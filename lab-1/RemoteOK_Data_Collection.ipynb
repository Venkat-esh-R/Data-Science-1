This notebook collects job postings data from the RemoteOK API and saves it into a CSV file.

```python

import requests
import pandas as pd
```

```python

# API endpoint
url = "https://remoteok.com/api"

# Get data from RemoteOK
response = requests.get(url)
if response.status_code != 200:
    raise Exception(f"Failed to fetch data: {response.status_code}")

data = response.json()

# Skip first element (metadata)
jobs_data = data[1:]
print(f"Total jobs fetched: {len(jobs_data)}")
```

```python

# Extract required fields
jobs_list = []
for job in jobs_data:
    jobs_list.append({
        "Company Name": job.get("company", ""),
        "Job Role": job.get("position", ""),
        "Location": job.get("location", ""),
        "Features/Tags": ", ".join(job.get("tags", []))
    })

# Create DataFrame
df = pd.DataFrame(jobs_list)
df.head()
```

```python

# Save the DataFrame to CSV
df.to_csv("remoteok_jobs.csv", index=False)
print("Data saved to remoteok_jobs.csv")
```
